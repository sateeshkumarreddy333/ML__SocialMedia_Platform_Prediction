IT DOES THIS BY MODELING THE RELATIONSHIP BETWEEN INDEPENDENT VARIABLES AND THE PROBABILITY OF BELONGING TO A SPECIFIC CLASS, USING A SIGMOID FUNCTION TO MAP THE OUTPUT TO A PROBABILITY BETWEEN 0 AND 1.***

```python
# Convert categorical columns to numeric using one-hot encoding
X_encoded = pd.get_dummies(x, drop_first=True)

# Now apply StandardScaler
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_scaled = sc.fit_transform(X_encoded)
X_scaled
```

***CLASSIFICATION REPORT***

```python
from sklearn.metrics import classification_report
target_names = y.unique()
print(classification_report(y_test, y_pred, target_names=target_names))
```

```python
print('Accuracy of the classifer is %0.2f' % metrics.accuracy_score(y_test,y_pred))
```

# ***MACHINE LEARNING MODEL: SUPPORT VECTOR MACHINE***


***SVM*** :
**A SUPPORT VECTOR MACHINE (SVM) IS A SUPERVISED MACHINE LEARNING ALGORITHM USED FOR BOTH CLASSIFICATION AND REGRESSION TASKS. ITS CORE PRINCIPLE INVOLVES FINDING A HYPERPLANE THAT OPTIMALLY SEPARATES DATA POINTS INTO DIFFERENT CLASSES WHILE MAXIMIZING THE MARGIN BETWEEN THE HYPERPLANE AND THE NEAREST DATA POINTS.**


#   *IMPORTING THE LIBRARY FILES*

```python
from sklearn.svm import SVC
```

#  ***SPLIT THE DATASET INTO TRAINING & TESTING***

```python
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.30)
```

##  ***BUILD THE MODEL(SVM***)

```python
from sklearn.preprocessing import LabelEncoder
# Assuming 'xtrain' is a pandas DataFrame and has categorical columns
label_encoder = LabelEncoder()

# Apply label encoding to all columns that contain non-numeric data
xtrain_encoded = xtrain.apply(lambda col: label_encoder.fit_transform(col) if col.dtype == 'object' else col)

# Initialize the SVC model
svc_model = SVC()  # You can customize this with parameters if needed

# Now fit the SVC model
svc_model.fit(xtrain_encoded, ytrain)
```

***DATA PREPROCESSING FOR MODEL PREDICTION***

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer

# Assuming your categorical columns are of type 'object'
categorical_columns = xtrain.select_dtypes(include=['object']).columns

# This will transform categorical columns into one-hot encoding
column_transformer = ColumnTransformer(
    transformers=[
        ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_columns)],
    remainder='passthrough')
# Create a pipeline to preprocess and fit the SVC model
pipeline = Pipeline(steps=[
    ('preprocessor', column_transformer),  # Apply the transformations
    ('svc', SVC())  # Apply SVC model after transformation
])
# Fit the model with the pipeline (this will automatically apply transformations)
pipeline.fit(xtrain, ytrain)
ypred = pipeline.predict(xtest)
```

```python
i = 0
print ("\n-------------------------------------------------------------------------")
print ('%-25s %-25s %-25s' % ('Original Label', 'Predicted Label', 'Correct/Wrong'))
print ("-------------------------------------------------------------------------")
for label in ytest:
 print ('%-25s %-25s' % (label, ypred[i]), end="")
 if (label == ypred[i]):
  print (' %-25s' % ('Correct'))
 else:
  print (' %-25s' % ('Wrong'))
 i = i + 1
print ("-------------------------------------------------------------------")
```

```python
print("\nClassification Report:\n",metrics.classification_report(ytest, ypred))
print ("---------------------------------------------------------------------")
```

```python
print('Accuracy of the classifer is %0.2f' % metrics.accuracy_score(ytest,ypred))
print ("-------------------------------------------------------------------")
```

# ***MACHINE LEARNING MODEL: K-NEAREST NEIGHBOR***


***K-NEAREST NEIGHBOR :
THE K-NEAREST NEIGHBORS (KNN) ALGORITHM IS A SIMPLE YET POWERFUL MACHINE LEARNING TECHNIQUE USED FOR BOTH CLASSIFICATION AND REGRESSION TASKS. IT PREDICTS THE CLASS OR VALUE OF A NEW DATA POINT BY LOOKING AT ITS "K" NEAREST NEIGHBORS IN THE TRAINING DATA. THE ALGORITHM CALCULATES DISTANCES BETWEEN THE NEW DATA POINT AND ALL POINTS IN THE TRAINING SET, SELECTS THE K CLOSEST NEIGHBORS, AND THEN EITHER ASSIGNS THE MOST FREQUENT CLASS LABEL***

#    *IMPORTING THE LIBRARY FILES*

```python
from sklearn.neighbors import KNeighborsClassifier
```

#  ***SPLIT THE DATASET INTO TRAINING & TESTING***

```python
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.30)
```

##   ***BUILD THE MODEL(K-NN***)

```python
# One-hot encode both training and test sets
Xtrain_encoded = pd.get_dummies(xtrain)
xtest_encoded = pd.get_dummies(xtest)

# Align test set to match train set's columns (fill missing columns with 0)
xtest_encoded = xtest_encoded.reindex(columns=xtrain_encoded.columns, fill_value=0)
```

***FIT THE MODEL AGAIN AND PREDICT***

```python
knn = KNeighborsClassifier(n_neighbors=5)
classifier = knn.fit(xtrain_encoded, ytrain)
ypred = classifier.predict(xtest_encoded)
```

***MODEL EVALUATION - RESULT COMPARISON***

```python
i = 0
print("\n-----------------------------------------------------------")
print('%-25s %-25s %-25s' % ('Original Label','Predicted Label','Correct/Worng'))
print("---------------------------------------------------------------")
for label in ytest:
  print('%-25s %-25s' % (label, ypred[i]), end="")
  if(label == ypred[i]):
    print('%-25s' %('correct'))
  else:
    print('%-25s' % ('Wrong'))
  i = i + 1
print("--------------------------------------------------------------")
```

***CLASSIFICATION REPORT***

```python
print("\nClassification Report:\n",metrics.classification_report(ytest, ypred))
print ("---------------------------------------------------------------------")
```

```python
print('Accuracy of the classifer is %0.2f' % metrics.accuracy_score(ytest,ypred))
print ("-------------------------------------------------------------------")
```
